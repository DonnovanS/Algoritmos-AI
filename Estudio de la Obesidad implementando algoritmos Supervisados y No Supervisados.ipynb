```python
original_df= df.copy()

# Se remplazan los valores por datos numericos
def change_descriptive_number(df, attr, arr_change):
  # toma el dataframe y remplaza el valor del atributo correspondiente con respecto
  # al arreglo de asignación
  df[attr]= df[attr].apply(lambda x: (np.where(arr_change == x))[0][0])
  return df

df = change_descriptive_number(df, 'Gender', genero)
df = change_descriptive_number(df, 'family_history_with_overweight', fh)
df = change_descriptive_number(df, 'CAEC', CAEC)
df = change_descriptive_number(df, 'CALC', CALC)
df = change_descriptive_number(df, 'MTRANS', MTRANS)
df = change_descriptive_number(df, 'SMOKE', SMOKE)
df = change_descriptive_number(df, 'SCC', SCC)
df = change_descriptive_number(df, 'FAVC', FAVC)
df
```

## <p style="background-color:DarkSlateGray;font-family:newtimeroman;color:#FFF9ED;font-size:120%;text-align:left;border-radius:10px 10px;">3. Aplicación de modelo de aprendizaje Supervisado, Maquina de vectores </p>


```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
'''
Aplicación de SVC, con pipeline para estandarizar los valore
se tiene encuenta que al realizar el proceso sin Standard StandardScaler
la variación de un 1% de efectividad
'''
cl = make_pipeline(StandardScaler(), SVC(C=7.0 , decision_function_shape='ovr', kernel='linear', gamma='scale', tol=0.2))
cl
```

```python
# entrenamiento del algoritmo
cl.fit(X_train, y_train)
```

```python
## Calculo de validación cruzada primaria
cl.score(X_test, y_test)
```

```python
# Predicción de acuerdo a los datos de entrenamiento
y_pred_svc= cl.predict( X_test )
unicos = df["NObeyesdad"].unique()
unicos
```

```python
# Obtención de metricas, Precisión, recall, f1-score, support
# Accuracy
from sklearn.metrics import accuracy_score, classification_report
y_pred = cl.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Verificación de metrica de mediante matriz de confusión
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true=y_test, y_pred=y_pred_svc)
cm_display = ConfusionMatrixDisplay(cm, display_labels=unicos).plot()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Eliminación de datos no representativos para la clasificación
df_clustering = df.drop(['NObeyesdad'], axis=1)
df_clustering_origin = df_clustering.copy()
# Elegir el número de Clusters
# Usar el método del codo para encontrar el número óptimo de clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_clustering)
    wcss.append(kmeans.inertia_)

# Graficar los resultados del método del codo
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
df_clustering.info()
```

```python
# Definición del modelo kmeans
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
# predicción del modelo
clusters= kmeans.fit_predict(df_clustering)
# número de clustering
cluster_labels = kmeans.labels_
# asignación de cluster al dataFrame
df_clustering['Cluster']= cluster_labels
```

## <p style="background-color:DarkSlateGray;font-family:newtimeroman;color:#FFF9ED;font-size:120%;text-align:left;border-radius:10px 10px;">5. Optimización del modelo del clustering</p>

```python
# Pipeline para realizar optimización del modelo
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
# Creación de pipeline para realizar una normalización de carateristicas
# y extracción de componentes principales
preprocessor = Pipeline(
       [
           ("scaler", MinMaxScaler()),
           ("pca", PCA(n_components=3, random_state=42)),
       ]
   )
```

```python
# creación de pipeline, con el algoritmo de clasificación kmenas,
kmean_cluster = Pipeline(
     [
         (
             "kmeans",
             KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0),
         ),
     ]
  )
pipe = Pipeline(
       [
           ("preprocessor", preprocessor),
           ("clusterer", kmean_cluster)
       ]
   )
pipe
```

```python
## entrenamiento y predicción
df_clustering_pipe= df_clustering_origin
clusters_pipe= pipe.fit_predict(df_clustering)
cluster_labels_pipe = pipe["clusterer"]["kmeans"].labels_
df_clustering_pipe['Cluster']= cluster_labels_pipe
df_clustering_pipe
```
