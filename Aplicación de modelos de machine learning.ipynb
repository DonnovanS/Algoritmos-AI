### **Carga de datos**

```python
#Realizamos la carga de directamente desde el link del .csv


# Reemplaza 'tu_archivo.csv' con la ruta de tu archivo
df = pd.read_csv('/content/data.csv', delimiter=';', decimal='.', encoding='utf-8')

# Limpieza de los nombres de las columnas si es necesario
df.columns = df.columns.str.replace('\t', '').str.strip('"')

# Muestra las primeras filas del DataFrame para verificar
df.head()

```

```python
df= df[df['Target'] != 'Enrolled']
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import LabelEncoder

# Identifica las columnas categóricas
columnas_categoricas = df.select_dtypes(include=['object']).columns

# Inicializa el LabelEncoder
le = LabelEncoder()

# Diccionario para mapear las etiquetas codificadas de vuelta a las etiquetas originales
mapeo_inverso = {}

# Aplica LabelEncoder a cada columna categórica
for col in columnas_categoricas:
    df[col] = le.fit_transform(df[col])
    mapeo_inverso[col] = {index: label for index, label in enumerate(le.classes_)}

# Muestra el mapeo inverso para cada columna
for col, mapping in mapeo_inverso.items():
    print(f"Mapping for {col}: {mapping}")
```

```python
#dividimos el dataset, en X todas las caracteristicas de los tumores y en Y la salida que corresponde a la clase que nos dice si el tumor es benigno o maligno
X = df1.to_numpy()[:, 0:36]
y = df1.to_numpy()[:, 36]
X.shape, y.shape
```

```python
# dividimos el conjunto de datos en subconjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
clf_NB = GaussianNB()
clf_NB.fit(X_train, y_train)
```

### **Clasificador Naive Bayes Gaussiano**

```python
#entrenamiento de un clasificador Naive Bayes gaussiano
from sklearn.naive_bayes import GaussianNB
clf_NB = GaussianNB()
clf_NB.fit(X_train, y_train)
tot_Test = np.size(y_test)
tot_aciertos = 0
y_pred = []
for x, y in zip(X_test, y_test):
    y_est = clf_NB.predict([x])
    y_pred.append(y_est)
    if (y==y_est): tot_aciertos +=1
    #print('Instancia %s Clase real %s - Clase estimada %s \n' %(x, y, y_est))
print('Exactitud del clasificador (porcentaje de aciertos) = %s \n' %(tot_aciertos*100/tot_Test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt


cm=confusion_matrix(y_test, y_pred)

# Matriz de confusión
print("Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))

# Informe de clasificación
print("\nInforme de Clasificación:")
print(classification_report(y_test, y_pred))

# Precisión
print("Precisión del Modelo:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true=y_test, y_pred=y_pred)
cm_display = ConfusionMatrixDisplay(cm, display_labels=['Dropout','Graduated'])
cm_display.plot(cmap='PuBu')
plt.title('Clasificador Naive Bayes Gaussiano', fontsize=18, loc='center', color='black', fontweight='bold')


```

### **Clasificación con Regresión Logística**









```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Crear un modelo de regresión logística
modelo_logistico = LogisticRegression()

# Entrenar el modelo en los datos de entrenamiento
modelo_logistico.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = modelo_logistico.predict(X_test)

# Calcular la precisión del modelo en el conjunto de prueba
precision = accuracy_score(y_test, y_pred)
print("Precisión del modelo de regresión logística:", precision)

# Generar la matriz de confusión y el informe de clasificación
matriz_confusion = confusion_matrix(y_test, y_pred)
informe_clasificacion = classification_report(y_test, y_pred)

# Imprimir la matriz de confusión y el informe de clasificación
print("Matriz de Confusión:")
print(matriz_confusion)
print("\nInforme de Clasificación:")
print(informe_clasificacion)
# Precisión
print("Precisión del Modelo:", accuracy_score(y_test, y_pred))
```
