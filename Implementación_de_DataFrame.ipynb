## **Pregunta orientadora**

***¿Cuál es el proceso para interpretar y manipular un data frame?***

El proceso para interpretar y manipular un DataFrame, según Galea y Capelo en "Applied Deep Learning with Python: Use Scikit-Learn, TensorFlow, and Keras to Create Intelligent Systems and Machine Learning Solutions" (2018), incluye los siguientes pasos:
1.	Carga de datos en un DataFrame de Pandas: Los datos pueden ser comúnmente almacenados en formatos CSV, TSV, SQL o JSON, posteriormente Pandas tiene la capacidad de leer estos formatos en un DataFrame.
2.	Crear un DataFrame: Una vez los datos estén disponibles, se procede a crear el DataFrame, por ejemplo si se tiene un arreglo de dos dimensiones se podría formar un DataFrame de la siguiente forma: pd.DataFrame(data=your_data, columns=your_column_names)
3.	Limpieza de datos: Se deben verificar y limpiar los datos antes del análisis, esto implica revisar valores faltantes o en error.
4.	Exploración de datos: La exploración de los datos se puede hacer mediante una descripción estadística para cada columna lo cual se realiza usando df.describe()
5.	Análisis de datos: Posteriormente se procede a realizar el análisis de datos requeridos o procesamiento
