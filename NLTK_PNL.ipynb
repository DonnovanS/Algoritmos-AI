```python
import requests
from bs4 import BeautifulSoup
import urllib.request

# URL del resumen de "Divina Comedia"
url = 'https://www.cervantesvirtual.com/obra-visor/la-voragine-989351/html/1e5af2ca-b590-4b72-9093-cf5c1cfb6ac8_2.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extracción del texto
text = ' '.join([p.get_text() for p in soup.find_all('p')])
print(text[:93])  # Mostrar los primeros 93 caracteres del texto
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Tokenización de palabras
tokens = word_tokenize(text.lower())
print(tokens[:20])  # Mostrar los primeros 20 tokens

# Eliminación de signos de puntuación
tokens_without_points = [word for word in tokens if word.isalpha()]
print(tokens_without_points[:20])

# Eliminación de stop words
stop_words = set(stopwords.words('spanish'))
cleans_tokens = [word for word in tokens_without_points if word not in stop_words]
print(cleans_tokens[:20])

```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Conteo de frecuencia de palabras
frequency_words = Counter(tokens_without_points)

# Visualización de las 10 palabras más comunes
palabras_comunes = frequency_words.most_common(10)
etiquetas, valores = zip(*palabras_comunes)

plt.figure(figsize=(10,5))
plt.bar(etiquetas, valores)
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.title('Las 10 palabras más comunes en "Divina Comedia"')
plt.show()
```

```python
# Verificar frecuencia de palabras
freq = nltk.FreqDist(tokens)
for key, val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
from nltk.corpus import wordnet

def synonyms(palabra, lang='eng'):
    synonym = []
    for syn in wordnet.synsets(palabra, lang=lang):
        for lemma in syn.lemmas(lang):
            synonym.append(lemma.name())
    return set(synonym)

# Ejemplo con la palabra "infierno" en español
synonyms_inf = synonyms('infierno', lang='spa') # Insertar una palabra en español para encontrar su sinónimo
print(synonyms_inf)
```
