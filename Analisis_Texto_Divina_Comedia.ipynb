```python
from bs4 import BeautifulSoup

import urllib.request
with urllib.request.urlopen('https://www.gradesaver.com/divina-comedia-infierno/guia-de-estudio/summary-cantos-1-3') as response:
    html = response.read()

soup = BeautifulSoup(html, "html5lib") # Limpia el texto HTML
text = soup.get_text(strip=True)
```

```python
#tokens
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text, language='spanish') # COLOQUE SU TEXTO AQUI Y COLOQUE EN ESPAÑOL
tokens = [word.lower() for word in tokens if word.isalpha()]
print(tokens)
```

```python
# Verificar frecuencia de palabras
freq_clean = nltk.FreqDist(clean_tokens)
for key, val in freq_clean.items():
    print(str(key) + ':' + str(val))
```

```python
from nltk.corpus import wordnet
synonyms = []
for syn in wordnet.synsets('infierno'): # AGREGUE UNA PALABRA PARA BUSCAR EL SINONIMO
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
print(synonyms)
```

```python
# Reemplazar tokens sinónimos
for ind, sin in enumerate(synonyms):
    clean_tokens_sin = [word.replace(synonyms[ind], 'independencia') for word in clean_tokens]
```

```python
# Reemplazar tokens sinónimos manual
sinonimos = ['infierno'] # ESTO ES UN EJEMPLO PERO PUEDE CAMBIAR POR LAS PALABRAS QUE QUIERA
for ind, sin in enumerate(sinonimos):
    clean_tokens_sin = [word.replace(sinonimos[ind], 'infierno') for word in clean_tokens]
```

```python
# Recalcular frecuencia de palabras con sinónimos agregados
freq_clean_sin = nltk.FreqDist(clean_tokens_sin)
for key, val in freq_clean_sin.items():
    print(str(key) + ':' + str(val))
```
