```python
import pandas as pd

nRowsRead = None
COLUMNS = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship'
           ,'race','sex','capital_gain','capital_loss','hours_week','native_country','label']
urlorgin = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df= pd.read_csv(urlorgin, delimiter=',',nrows = nRowsRead, names=COLUMNS)
df.dataframeName = 'house-votes-84.data.csv'
nRow, nCol = df.shape
print(f'Hay {nRow} filas y {nCol} columnas')
```

```python
#Muestra cabecera del dataframe en n columnas
df.head(5)
```

```python
#Conteo de datos
df.shape
```

```python
#Elimina filas no deseadas
df.drop(['label'], axis = 'columns', inplace=True)
```

```python
df.dtypes
```

```python
#Eliminar datos que contengan un espacio
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
```

```python
# Utiliza applymap para verificar la presencia del carácter '?' en el dataframe
total_characters = df.applymap(lambda n: str(n).count('?')).values.sum()
print("Celdas con un valor igual a '?:", total_characters)
```

```python
df_new = df[~(df == '?').any(axis=1)]
df_new
```

```python
df_new.dropna()
```
