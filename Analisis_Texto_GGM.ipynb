```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')  # Descargar recursos adicionales para WordNe
```

```python
from bs4 import BeautifulSoup
import urllib.request

with urllib.request.urlopen('https://www.culturagenial.com/es/cien-anos-de-soledad-de-gabriel-garcia-marquez/') as response:
    html = response.read()

soup = BeautifulSoup(html, "html5lib")
text = soup.get_text(strip=True)
print(text[:1000])  # Imprimir los primeros 1000 caracteres para verificar la lectura
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text, language='spanish')
tokens = [word.lower() for word in tokens if word.isalpha()]
print(f"Tokens: {tokens[:50]}")  # Mostrar los primeros 50 tokens
```

```python
tokens = [word.lower() for word in tokens if word.isalpha()]
print(f"Tokens sin puntuación: {tokens[:50]}")  # Mostrar los primeros 50 tokens sin puntuación
```

```python
freq = nltk.FreqDist(tokens)
for key, val in freq.items():
    print(f"{key}: {val}")

# Visualizar tokens
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

freq.plot(30, cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('spanish'))
clean_tokens = [token for token in tokens if token not in stop_words]
print(f"Clean Tokens: {clean_tokens[:50]}")  # Mostrar los primeros 50 tokens limpios

# Verificar la frecuencia de palabras sin stop words
freq_clean = nltk.FreqDist(clean_tokens)
for key, val in freq_clean.items():
    print(f"{key}: {val}")
```

```python
from nltk.corpus import wordnet

synonyms = []
for syn in wordnet.synsets('soledad', lang='spa'):
    for lemma in syn.lemmas('spa'):
        synonyms.append(lemma.name())
print(f"Sinónimos de 'soledad': {synonyms}")
```

```python
clean_tokens_sin = clean_tokens.copy()
for syn in synonyms:
    clean_tokens_sin = [word.replace(syn, 'independencia') for word in clean_tokens_sin]

# Reemplazo manual de tokens por sinónimos
sinonimos_manual = ['libertad', 'independencia']
for sin in sinonimos_manual:
    clean_tokens_sin = [word.replace(sin, 'libertad') for word in clean_tokens_sin]

# Recalcular frecuencia de palabras con sinónimos agregados
freq_clean_sin = nltk.FreqDist(clean_tokens_sin)
for key, val in freq_clean_sin.items():
    print(f"{key}: {val}")

# Visualizar tokens con sinónimos
freq_clean_sin.plot(30, cumulative=False)
plt.show()

```
