#### 1. Importar librerías

```python
# Importar librerías

from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

# no mostrar advertencias
import warnings
warnings.filterwarnings("ignore")
```

#### 2. Crear dataframe para almacenar información

```python
columns=['Película_ID','Titulo','Año','Genero']
df = pd.DataFrame(columns=columns)
df
```

#### 3. Web Scrapping desde tablas de wikipedia

* Para este proceso se indicarán las URL del texto a obtener y luego se creará una función que busque la tabla requerida y sus datos.

```python

# objeto / lista urls donde se indicará las direcciones de las páginas de interés
URL = {2000:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2000',
       2001:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2001',
       2002:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2002',
       2003:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2003',
       2004:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2004',
       2005:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2005',
       2006:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2006',
       2007:'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2007'
      }
id = 34759

# para cada url en la lista
for key, url in URL.items():
    # establecer requerimiento y obtener respuesta
    response = requests.get(url)
    # crear "sopa" árbol de objetos
    soup = BeautifulSoup(response.text, 'html.parser')
    # definir búsqueda de tabla en la sopa
    table = soup.findAll('table', {'class':'wikitable'})[1].tbody
    # número de filas de la tabla a obtener (todas)
    rows = table.find_all('tr')
    # para cada i en el rango
    for i in range(1, len(rows)):
        # objeto para almacenar las filas de tabla seleccionadas
        tds = rows[i].find_all('td')
        # registrar valores en lista
        values = [id,tds[0].text, key, tds[3].text]
        id+=1
        # colectar en el dataframe creado
        df = df.append(pd.Series(values, index=columns), ignore_index=True)

# mostrar dataframe
df
```

#### Guardar Dataframe en formato .csv

```python
df.to_csv('Peliculas.csv', header= True, index=False)
```
