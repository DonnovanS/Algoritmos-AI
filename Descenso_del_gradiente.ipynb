**El descenso del gradiente se utiliza para alcanzar el mínimo de la función de error(se produce al dar los valores iniciales a los pesos de la neuronas de la red neuronal y el bias), para actualizar los pesos y el bias o sesgo en el entrenamiento de la red neuronal profunda artificial**

```python
# Importamos los módulos a usar.
import numpy as np
import matplotlib.pyplot as plt # importa la libreria matplotlib con el alias plt
from numpy.random import randint # importa la función randint de la librería numpy.random
from numpy import linspace # importa la función linspace de la librería numpy
import pandas as pd   # importa la libreria pandas con el alias pd

%matplotlib inline
```

```python
# Creamos la función a optimizar y la grafica.
# lambda es un tipo de funcion llamada anonima.

funcion = lambda x: x**2+ 1
# se crean 200 valores espaciados entre x=-10 y x=10
x = np.linspace(-10,10,200)
# Grafica la función.
plt.plot(x, funcion(x))
plt.show()
```

```python
# Creamos la función de la derivada de x.
# utilizamos def nombredelafuncion():
# return=valor de retorno: resultado
def derivada(x):

    x_derivada = 2*x
    return x_derivada # Regresa la derivada de la función inicial

# Creamos la función del gradiente
def gradiente(x_inicio, precision, l_r):

# se crean dos listas para los valores de x,y
#se utilizaran despues para la visualizacion puntual.



    x_list, y_list = [x_inicio], [funcion(x_inicio)]

    # Utilizamos el bucle While para hacer el recorrido

    while True:

        '''
         Se Obtiene el valor de la pendiente de la derivada
        debido a que necesitamos un descenso negativo (hacia el mínimo)
        usamos el signo menos '-'

        '''

        d_x = - derivada(x_inicio)

        # Obtenemos el nuevo valor de x sumando el valor previo de x más la multiplicación del l_r por la derivada.
        x_inicio = x_inicio + (l_r * d_x)

        # Agregamos  o añadimos el nuevo valor de x a la lista para posteriormente visualizarlo.
        # metodo append() para agregar un valor a la lista
        x_list.append(x_inicio)

        # Agregamos o añadimos el nuevo valor de y a la lista para posteriormente visualizarlo.
        y_list.append(funcion(x_inicio))


        #abs significa valor absoluto
        # x_list[-1] significa última valor en la lista
        # x_list[-2] significa valor antes del último
        # abs(x_list[-1] - x_list[-2]) <= precision.
        #significa si la resta de los dos es menor que la precision pare.
        # Salimos del bucle cuando se cumpla la condición.

        if abs(x_list[-1] - x_list[-2]) <= precision:
            break

    # Visualizamos los resultados.
    # str() convierte a string.
    # len() longitud de la lista.

    print ("El mínimo local se produce en: "+ str(x_inicio))

    # Número de pasos para obtener el minimo local.

    print ("Número de pasos: " + str(len(x_list)))

    # subplot para ver varias graficas en una.
    # visualizar la grafico de dispersion mediante plt.scatter()

    plt.subplot(1,2,2)# tambien se puede escribir plt.subplot(122)
    plt.scatter(x_list,y_list,c="b")# color=c="b" significa b=blue color azul,y= yellow(amarillo)
    plt.plot(x_list,y_list,c="b")# gráfico para ver el recorrido que realizo el gradiente para obtener el minimo.
    plt.plot(x,funcion(x), c="y")# gráfico de la funcion inicial
    plt.title("Gráfico del descenso del gradiente paso a paso.")
    plt.show()
```

**LLamamos la función gradiente pasando los parametros, punto inicial de x,precision y learning rate(tasa de aprendizaje)**

```python
gradiente(-8, 0.0001, 0.05)
```

**Varien esos 3 valores del Descenso del Gradiente**

```python
gradiente(-8, 0.01, 0.03)
```

```python
gradiente(5, 0.01, 0.03)
```

```python
import numpy as np
import matplotlib.pyplot as plt # importa la libreria matplotlib con el alias plt
from numpy.random import randint # importa la función randint de la librería numpy.random
from numpy import linspace # importa la función linspace de la librería numpy
import pandas as pd   # importa la libreria pandas con el alias pd
# Creamos la función a optimizar y la grafica.
# lambda es un tipo de funcion llamada anonima.
%matplotlib inline

funcion2 = lambda x: x**2+ 5*x + 6
# se crean 200 valores espaciados entre x=-10 y x=10
x2 = np.linspace(-10,10,200)
# Grafica la función.
plt.figure(figsize=(8, 6))
plt.plot(x2, funcion2(x2))
plt.show()
```

```python
# Creamos la función de la derivada de x.

def derivada(x):

    x2_derivada = 2*x+5
    return x2_derivada # Regresa la derivada de la función inicial

# Creamos la función del gradiente
def gradiente(x2_inicio, precision2, l_r2):

# se crean dos listas para los valores de x,y
#se utilizaran despues para la visualizacion puntual.


    x2_list, y2_list = [x2_inicio], [funcion2(x2_inicio)]

    # Utilizamos el bucle While para hacer el recorrido

    while True:

        d_x2 = - derivada(x2_inicio)

        # Obtenemos el nuevo valor de x sumando el valor previo de x más la multiplicación del l_r por la derivada.
        x2_inicio = x2_inicio + (l_r2 * d_x2)

        # Agregamos  o añadimos el nuevo valor de x a la lista para posteriormente visualizarlo.
        # metodo append() para agregar un valor a la lista
        x2_list.append(x2_inicio)

        # Agregamos o añadimos el nuevo valor de y a la lista para posteriormente visualizarlo.
        y2_list.append(funcion2(x2_inicio))


        if abs(x2_list[-1] - x2_list[-2]) <= precision2:
            break

    print ("El mínimo local se produce en: "+ str(x2_inicio))

    print ("Número de pasos: " + str(len(x2_list)))

    # subplot para ver varias graficas en una.
    # visualizar la grafico de dispersion mediante plt.scatter()

    plt.subplot(1,2,2)# tambien se puede escribir plt.subplot(122)
    plt.scatter(x2_list,y2_list,c="r")# color=c="b" significa b=blue color azul,y= yellow(amarillo)
    plt.plot(x2_list,y2_list,c="r")# gráfico para ver el recorrido que realizo el gradiente para obtener el minimo.
    plt.plot(x2,funcion2(x2))# gráfico de la funcion inicial
    plt.title("Gráfico del descenso del gradiente paso a paso.")
    plt.show()
```
